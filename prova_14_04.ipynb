Par ou ímpar
Peça um número ao usuário e informe se ele é par ou ímpar.

```python
numero = int(input(" número: "))

if numero % 2 == 0:
    print(f"O número {numero} é par.")
else:
    print(f"O número {numero} é ímpar.")
```

Maior de dois números
Solicite dois números e mostre qual é o maior (ou se são iguais).



```python
numero1 = int(input("número um: "))
numero2 = int(input("número dois: "))

if numero1 > numero2:
    print(f"O primeiro número ({numero1}) é maior que o segundo número ({numero2}).")
elif numero2 > numero1:
    print(f"O segundo número ({numero2}) é maior que o primeiro número ({numero1}).")
else:
    print(f"Os números são iguais ({numero1} e {numero2}).")
```

Aprovação de aluno
Peça a nota de um aluno e informe:
Aprovado (nota ≥ 7)
Recuperação (5 ≤ nota < 7)
Reprovado (nota < 5)



```python
nota = float(input("nota do aluno: "))

if nota >= 7:
    print("Aprovado")
elif nota >= 5 and nota < 7:
    print("Recuperação")
else:
    print("Reprovado")
```

Classificação de idade
Receba a idade e classifique:
Criança (0–12)
Adolescente (13–17)
Adulto (18–59)
Idoso (60+)


```python
idade = int(input("idade: "))

if 0 <= idade <= 12:
    print("Criança")
elif 13 <= idade <= 17:
    print("Adolescente")
elif 18 <= idade <= 59:
    print("Adulto")
elif idade >= 60:
    print("Idoso")
```

Calculadora simples
Peça dois números e uma operação (+, -, *, /). Use if/elif para calcular o resultado.


```python
numero1 = float(input(" número um: "))
numero2 = float(input(" número dois: "))
operacao = input("Digite a operação (+, -, *, /): ")

if operacao == '+':
    resultado = numero1 + numero2
    print(f"Resultado: {numero1} + {numero2} = {resultado}")
elif operacao == '-':
    resultado = numero1 - numero2
    print(f"Resultado: {numero1} - {numero2} = {resultado}")
elif operacao == '*':
    resultado = numero1 * numero2
    print(f"Resultado: {numero1} * {numero2} = {resultado}")
elif operacao == '/':
    if numero2 != 0:
        resultado = numero1 / numero2
        print(f"Resultado: {numero1} / {numero2} = {resultado}")
    else:
        print("Erro: Divisão por zero não é permitida.")
else:
    print("Operação inválida. Por favor, use +, -, *, ou /.")
```

Dia da semana
Peça um número de 1 a 7 e mostre o dia correspondente (ex: 1 = domingo).



```python
numero_dia = int(input("Digite um número de 1 a 7 para o dia da semana: "))

dias_da_semana = {
    1: "Domingo",
    2: "Segunda-feira",
    3: "Terça-feira",
    4: "Quarta-feira",
    5: "Quinta-feira",
    6: "Sexta-feira",
    7: "Sábado"
}

if numero_dia in dias_da_semana:
    print(f"O dia correspondente é: {dias_da_semana[numero_dia]}")
else:
    print("Número inválido. Por favor, digite um número entre 1 e 7.")
```

Ano bissexto
Solicite um ano e informe se ele é bissexto ou não.

```python
ano = int(input("Digite um ano: "))

if (ano % 4 == 0 and ano % 100 != 0) or (ano % 400 == 0):
    print(f"O ano {ano} é bissexto.")
else:
    print(f"O ano {ano} não é bissexto.")
```

Desconto em compras
Peça o valor de uma compra:
≥ 100: 10% de desconto
≥ 50 e < 100: 5% de desconto
< 50: sem desconto


```python
valor_compra = float(input("valor da compra: "))

if valor_compra >= 100:
    desconto = valor_compra * 0.10
    valor_final = valor_compra - desconto
    print(f"Valor da compra: R${valor_compra:.2f}")
    print(f"Desconto aplicado (10%): R${desconto:.2f}")
    print(f"Valor final com desconto: R${valor_final:.2f}")
elif valor_compra >= 50 and valor_compra < 100:
    desconto = valor_compra * 0.05
    valor_final = valor_compra - desconto
    print(f"Valor da compra: R${valor_compra:.2f}")
    print(f"Desconto aplicado (5%): R${desconto:.2f}")
    print(f"Valor final com desconto: R${valor_final:.2f}")
else:
    print(f"Valor da compra: R${valor_compra:.2f}")
    print("Sem desconto aplicado.")
    print(f"Valor final: R${valor_compra:.2f}")
```
