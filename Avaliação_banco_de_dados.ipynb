Questão 3 – Verificação de triângulo

```python
# Entrada
lado1 = float(input("Digite o valor do primeiro lado: "))
lado2 = float(input("Digite o valor do segundo lado: "))
lado3 = float(input("Digite o valor do terceiro lado: "))

# Processamento e Saída
if (lado1 + lado2 > lado3) and (lado1 + lado3 > lado2) and (lado2 + lado3 > lado1):
  print("Os valores podem formar um triângulo válido.")
else:
  print("Os valores NÃO podem formar um triângulo válido.")
```

Questão 6 – Cálculo de salário com bônus

```python
# Entrada
salario = float(input("Digite o salário do funcionário: "))
pontuacao_desempenho = int(input("Digite a pontuação de desempenho (0 a 10): "))

# Processamento - Cálculo do bônus
bonus_percentual = 0
if pontuacao_desempenho < 5:
  bonus_percentual = 0  # Sem bônus
elif pontuacao_desempenho >= 5 and pontuacao_desempenho <= 7:
  bonus_percentual = 0.05 # 5% de bônus
elif pontuacao_desempenho >= 8 and pontuacao_desempenho <= 9:
  bonus_percentual = 0.10 # 10% de bônus
elif pontuacao_desempenho == 10:
  bonus_percentual = 0.15 # 15% de bônus
else:
  print("Pontuação de desempenho inválida. Por favor, digite um valor entre 0 e 10.")

valor_bonus = salario * bonus_percentual
salario_final = salario + valor_bonus

# Saída
print(f"Valor do bônus: R$ {valor_bonus:.2f}")
print(f"Salário final com bônus: R$ {salario_final:.2f}")
```
