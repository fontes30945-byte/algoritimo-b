# Escrever um programa que o usuario digita um numero 1 a 10 e o programa imprimi a tabuada do numero digitado

```python
numero = int(input("Di1gite um número de 1 a 10 para ver sua tabuada: "))

if 1 <= numero <= 10:
  print(f"Tabuada do {numero}:")
  for i in range(1, 11):
    resultado = numero * i
    print(f"{numero} x {i} = {resultado}")
else:
  print("Por favor, digite um número entre 1 e 10.")
```
